# **Import Packages**

```python
import geopandas as gpd
```

# **Dataset#1**

```python
#loading dataset in
df1 = gpd.read_file('Cancer_Rates.geojson')
df1
```

```python
# missing values
missing_values = df1.isnull().sum()
missing_values
```

```python
# column names
df1.columns
```

```python
# map of breast cancer for Lake County Illinois
df1.plot("Breast_Can", legend=True)
```

```python
# map of ALL CANCER for Lake County Illinois
df1.plot("All_Cancer", legend=True)
```

### **Summary**
This [dataset](https://catalog.data.gov/dataset/cancer-rates-5cf0c) contains information on cancer Rates for Lake County Illinois. Specifically, I chose to focus on Breast Cancer for the first visualization and all cancers for the second visualization compare the amount of a specific cancer to all cancer in Illinois. The lighter the shade the more instances of cancer.

Based on the breast cancer, southeastern part of Illinois has a greater instance of breast cancer. However, in terms of all cancer, the second visualization shows that only a specific part of the southeast has both the greatest instance of breast cancer AND all cancer.
